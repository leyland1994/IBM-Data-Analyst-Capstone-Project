<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv")

# Clean the 'ConvertedCompYearly' column by removing rows with missing values
df_cleaned_comp = df.dropna(subset=['ConvertedCompYearly', 'Country'])

# Identify countries with a significant number of respondents for better comparison
country_counts = df_cleaned_comp['Country'].value_counts()
top_countries = country_counts[country_counts >= 500].index.tolist()
df_top_countries = df_cleaned_comp[df_cleaned_comp['Country'].isin(top_countries)]

# Create box plots to compare compensation distributions across the top countries
plt.figure(figsize=(16, 10))
sns.boxplot(x='Country', y='ConvertedCompYearly', data=df_top_countries)
plt.title('Compensation Distribution Across Top Countries')
plt.xlabel('Country')
plt.ylabel('Annual Compensation (USD)')
plt.xticks(rotation=45, ha='right')
plt.yscale('log') # Using a logarithmic scale to handle potential outliers and wide ranges
plt.tight_layout()
plt.show()

# Analyze compensation range and potential anomalies
compensation_stats = df_cleaned_comp.groupby('Country')['ConvertedCompYearly'].agg(['min', 'max', 'median', 'mean', 'std', 'count'])
print("\nCompensation Statistics by Country (for countries with >= 500 respondents):")
print(compensation_stats.loc[top_countries].sort_values(by='median', ascending=False))

# Further analysis on potential anomalies (example: extremely high values within a country)
def identify_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series[(series < lower_bound) | (series > upper_bound)]

print("\nExamples of potential high compensation anomalies (outliers) by country:")
for country in top_countries:
    outliers = identify_outliers(df_top_countries[df_top_countries['Country'] == country]['ConvertedCompYearly'])
    if not outliers.empty:
        print(f"\n{country}:")
        print(outliers.head()) # Displaying a few top outliers
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here
import pandas as pd

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv")

# Handle missing values in 'ConvertedCompYearly' by dropping rows
df_cleaned_comp = df.dropna(subset=['ConvertedCompYearly'])

# Calculate the IQR for 'ConvertedCompYearly'
Q1 = df_cleaned_comp['ConvertedCompYearly'].quantile(0.25)
Q3 = df_cleaned_comp['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower bounds for outlier removal
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new DataFrame without outliers in 'ConvertedCompYearly'
df_no_comp_outliers = df_cleaned_comp[
    (df_cleaned_comp['ConvertedCompYearly'] >= lower_bound) &
    (df_cleaned_comp['ConvertedCompYearly'] <= upper_bound)
]

# Print the shape of the original cleaned DataFrame and the new DataFrame
print(f"Shape of DataFrame after handling missing 'ConvertedCompYearly': {df_cleaned_comp.shape}")
print(f"Shape of DataFrame after removing 'ConvertedCompYearly' outliers: {df_no_comp_outliers.shape}")

# Display descriptive statistics of 'ConvertedCompYearly' before and after outlier removal
print("\nDescriptive Statistics of 'ConvertedCompYearly' before outlier removal:")
print(df_cleaned_comp['ConvertedCompYearly'].describe())

print("\nDescriptive Statistics of 'ConvertedCompYearly' after outlier removal:")
print(df_no_comp_outliers['ConvertedCompYearly'].describe())

# The 'df_no_comp_outliers' DataFrame is now ready for correlation analysis
# as it contains the data with outliers removed from the 'ConvertedCompYearly' column.
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv")

# Clean and prepare the data for correlation analysis
# Drop rows with missing values in the relevant columns
df_cleaned = df.dropna(subset=['ConvertedCompYearly', 'WorkExp', 'JobSat'])

# Map 'JobSat' to numerical values
job_sat_mapping = {'Very dissatisfied': 1, 'Dissatisfied': 2, 'Neutral': 3, 'Satisfied': 4, 'Very satisfied': 5}
df_cleaned['JobSatPoints_1'] = df_cleaned['JobSat'].map(job_sat_mapping)

# Select the columns for correlation
df_corr = df_cleaned[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']]

# Calculate the correlation matrix
correlation_matrix = df_corr.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Compensation, Work Experience, and Job Satisfaction')
plt.tight_layout()
plt.show()

# Print the correlation matrix for numerical output
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv")

# Clean and prepare the data
df_cleaned = df.dropna(subset=['ConvertedCompYearly', 'WorkExp', 'JobSat'])
job_sat_mapping = {'Very dissatisfied': 1, 'Dissatisfied': 2, 'Neutral': 3, 'Satisfied': 4, 'Very satisfied': 5}
df_cleaned['JobSatPoints_1'] = df_cleaned['JobSat'].map(job_sat_mapping)

# --- Scatter plot: ConvertedCompYearly vs. WorkExp ---
plt.figure(figsize=(10, 6))
sns.scatterplot(x='WorkExp', y='ConvertedCompYearly', data=df_cleaned)
plt.title('Scatter Plot: Annual Compensation vs. Work Experience')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Annual Compensation (USD)')
plt.yscale('log')  # Using a logarithmic scale for better visualization of compensation range
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Scatter plot: ConvertedCompYearly vs. JobSatPoints_1 ---
plt.figure(figsize=(10, 6))
sns.scatterplot(x='JobSatPoints_1', y='ConvertedCompYearly', data=df_cleaned)
plt.title('Scatter Plot: Annual Compensation vs. Job Satisfaction')
plt.xlabel('Job Satisfaction (1: Very Dissatisfied to 5: Very Satisfied)')
plt.ylabel('Annual Compensation (USD)')
plt.yscale('log')  # Using a logarithmic scale for better visualization of compensation range
plt.xticks(range(1, 6))  # Ensure x-axis ticks are at the job satisfaction levels
plt.grid(True)
plt.tight_layout()
plt.show()
```
